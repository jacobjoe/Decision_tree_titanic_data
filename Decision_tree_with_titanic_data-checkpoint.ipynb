# Decision Tree

```python
#Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv('K:\henry\python\data\Titanic for Decision tree.csv')
data.head()
```

```python
# Check missing value
data.isnull().sum()
```

```python
# Visually check missing value 
sns.heatmap(data.isnull())
```

```python
data.head(3)
```

```python
# Seperate independent and dependent variables
y = data['Survived']
X = data.drop(['Survived'], axis = 1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Create train and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42, stratify = y)
```

```python
# Distribution in training set
y_train.value_counts(normalize= True)
```

```python
# Distribution in testing set
y_test.value_counts(normalize= True)
```

```python
X_train.shape, y_train.shape
```

```python
X_test.shape, y_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
# Create Decision tree function
dt_model = DecisionTreeClassifier(random_state=10)
```

```python
# Fit the model
dt_model.fit(X_train, y_train)
```

```python
# Check the train score
dt_model.score(X_train, y_train)
```

```python
# Check the test score
dt_model.score(X_test, y_test)
```

```python
# Prediction on test set
dt_model.predict(X_test)
```

```python
# Check the probability of prediction
dt_model.predict_proba(X_test)
```

```python
# Probability of prediction on X_test set and we take only 1st column
y_pred = dt_model.predict_proba(X_test)[:, 1]
```

```python
# Fix a threshold value 0.5 and y_pred values are change to 0's and 1's
new_y = []
for x in range(len(y_pred)):
    if y_pred[x] < 0.6:
        new_y.append(0)
    else:
        new_y.append(1)
```

```python
from sklearn.metrics import accuracy_score
```

```python
# Check the accuracy score
accuracy_score(y_test, new_y)
```

# Change the max depth

```python
# Use some parameters from DecisionTreeClassifier and optimize performance of decision tree
training_acc = []
testing_acc = []
for x in range(1, 11):
    dt_model = DecisionTreeClassifier(random_state=10, max_depth=x)
    dt_model.fit(X_train, y_train)
    training_acc.append(dt_model.score(X_train, y_train))
    testing_acc.append(dt_model.score(X_test, y_test))
```

```python
# Create a dataframe from previous output
df = pd.DataFrame({'Max depth':range(1, 11), 'Train acc':training_acc, 'Test acc':testing_acc})
df.head()
```

```python
# Plot the values 
plt.rcParams['figure.figsize'] = (15, 15)
plt.plot(df['Max depth'], df['Train acc'], c = 'r', label = 'Train acc', marker = 'o')
plt.plot(df['Max depth'], df['Test acc'], c = 'b', label = 'Test acc', marker = 'o')
plt.title('Graph')
plt.xlabel('Depth of tree')
plt.ylabel('Performance')
plt.legend()
plt.show()
```

```python
# Use some other parameter and set the max_depth based on the above figure
dt_model = DecisionTreeClassifier(random_state=10, max_depth=6, max_leaf_nodes=25)
```

```python
dt_model.fit(X_train, y_train)
```

```python
dt_model.score(X_train, y_train)
```

```python
dt_model.score(X_test, y_test)
```
